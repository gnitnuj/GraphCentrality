```python
import networkx as nx
%matplotlib inline
```

```python
M=nx.read_gml("polblogs.gml")
```

```python
M
```

```python
# create weighted graph from M
G = nx.Graph()
for u,v,data in M.edges_iter(data=True):
    if G.has_edge(u,v):
        G[u][v]['weight'] += 1
    else:
        G.add_edge(u, v, weight=1)
```

```python
G
```

### 1.  Create a nice looking graph of the political blog network

```python
nx.draw(G)
```

```python
# prettier graph ?
pos = nx.spring_layout(G)
nx.draw_networkx_labels(G,pos)
nx.draw(G,pos,node_color="w")
```

### 2.  What are the 10 most influential blogs?

```python
centrality = nx.eigenvector_centrality_numpy(G)
```

```python
centrality
```

```python
# get the key with the max value
max(centrality, key=centrality.get)
```

Going by eigenvector centrality, the most influential blog is...

atrios.blogspot.com

### 3.  What are the 10 least influential blogs?



```python
# get the key with the min value
min(centrality, key=centrality.get)
```

Going by eigenvector centrality, the least influential blog is...

thelonedem.com
